```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2*(np.pi), 0.1)
y = np.sin(x)
z = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y, color='red')
plt.xlabel("x")
plt.ylabel("sinx")
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,z, color='green')
plt.xlabel("x")
plt.ylabel("cosx")
plt.title("Cosine")

plt.suptitle("SINE AND COSINE", fontweight ="bold")
plt.tight_layout()
plt.show()
```

```python
# (a) Line plot

fig = plt.figure(figsize=(12,12))

data1 = np.random.randint(1, 100, 20)

fig.add_subplot(3,2,1)
plt.plot(data1, color='orange')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("LINE PLOT", fontweight ="bold")

# (b) Bar plot
data2 = np.arange(10)
data3 = np.random.randint(5, size=(10))

fig.add_subplot(3,2,2)
plt.bar(data2, data3, color='pink', width=0.5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("BAR PLOT", fontweight ="bold")

# (c) Scatter plot
data4 = np.arange(50)
data5 = np.random.randn(50)

fig.add_subplot(3,2,3)
plt.scatter(data4, data5, color='black')
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("SCATTER PLOT", fontweight ="bold")

# (d) Pie plot
names = ['one', 'two', 'three', 'four', 'five', 'six']
data = np.random.randint(100, size=6)

fig.add_subplot(3,2,4)
plt.pie(data, labels=names, shadow=True)
plt.legend(title='names', loc='upper right', bbox_to_anchor=(1.4, 1.0))
plt.title("PIE PLOT", fontweight ="bold")

# (e) Area plot
data6 = range(1,11)
data7 = np.random.randint(100, size=10)

fig.add_subplot(3,2,5)
plt.fill_between(data6, data7, color='red')
plt.title("AREA PLOT", fontweight ="bold")

# (f) Histogram
np.random.seed(10**7)
n_bins = 20
x = np.random.randn(10000, 3)
colors = ['green', 'blue', 'lime']

fig.add_subplot(3,2,6)
plt.hist(x, n_bins, density = True, histtype ='bar', color = colors, label = colors)
plt.legend(prop ={'size': 10})
plt.title('HISTOGRAM', fontweight ="bold")

plt.tight_layout()
plt.show()
```
